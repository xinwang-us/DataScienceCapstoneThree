Build RFM table for RFM Customer Segmentation and K-Means Clustering

```python
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uk_online_retail = pd.read_excel('data\\new_uk_online_retail.xlsx')
uk_online_retail.head()
```

```python
uk_online_retail.drop('Unnamed: 0', axis=1, inplace = True)
uk_online_retail.head()
```

```python
cut_off_date = uk_online_retail.InvoiceDay.max()+timedelta(days=1)
cut_off_date
```

```python
RFM_data = uk_online_retail.groupby('CustomerID').agg({'InvoiceDay': lambda x: (cut_off_date - x.max()).days, \
                                                       'DayOrderCount': 'sum', \
                                                       'DayOrderAmount': 'sum'})
RFM_data.head()
```

```python
RFM_data.rename(columns={'InvoiceDay': 'Recency', 'DayOrderCount': 'Frequency', 'DayOrderAmount':'MonetaryValue'}, inplace=True)
RFM_data.head()
```
