```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data\Online Retail.xlsx')
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.info()
```

```python
# remove rows with null Customer IDs
df = df[df['CustomerID' ].notna()]
df.info()
```

```python
#convert Costomer ID data type from float to integer
df.loc[:, 'CustomerID'] = df['CustomerID'].astype('int')
df.dtypes
```

```python
df.head()
```

```python
df['Country'].value_counts()
```

```python
df[df.Quantity<=0]
```

```python
df[df.InvoiceNo.str.startswith('C') == 1]
```

```python
#check whether the invoice numbers always start with 'C' wherever the negative quantities occur
set(df[df.InvoiceNo.str.startswith('C') == 1].index) - set(df[df.Quantity<=0].index)
```

```python
set(df[df.Quantity<=0].index) - set(df[df.InvoiceNo.str.startswith('C') == 1].index)
```

Conclusion: all the rows with negative quantity have the Invoices start with 'C' which may stand for cancelation

```python
negative_quantity_df = df[df.Quantity<=0].copy()
negative_quantity_df.head()
```

```python
negative_quantity_df.tail()
```

```python
negative_quantity_df.InvoiceNo.str.len().unique()
```

```python
negative_quantity_df['InvoiceNum'] = negative_quantity_df.InvoiceNo.str.slice(1).astype('int')
negative_quantity_df.dtypes
```

```python
negative_quantity_df.head()
```

```python
positive_quantity_df = df[df.Quantity > 0].copy()
positive_quantity_df.shape, positive_quantity_df.dtypes
```

```python
positive_quantity_df['InvoiceNum'] = positive_quantity_df.InvoiceNo.astype('int')
positive_quantity_df.dtypes
```

```python
df[df['CustomerID']==16321]
```

Observations for orders from the Customer with ID 16321: The cancelations came before or at the same time when the orders were placed.

```python
df[df['CustomerID']==12668].tail(60)
```

Observations for orders from the Customer with ID 12668: NO orders were placed before cancelation.

```python
set(positive_quantity_df['InvoiceNum']).intersection(set(negative_quantity_df['InvoiceNum'])) 
```

Conclusion: No orders were placed before order cancelation. So, we'll remove all the rows with negative quantity.

```python
positive_quantity_df.drop('InvoiceNo', axis=1, inplace=True)
positive_quantity_df.columns
```

```python
positive_quantity_df.to_excel('data\online_retail_clean.xlsx')
```

```python
uk_online_retail_clean = positive_quantity_df[positive_quantity_df.Country=='United Kingdom']
uk_online_retail_clean.shape
```
